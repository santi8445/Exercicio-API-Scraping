```python
import requests
import pandas as pd
```

```python
#Comprueba si España está entre los paises que tienen estaciones de AirVisual
```

```python
key = '710d13ad-e9e3-4f22-bf06-5720cc89ccc2'
url = f'http://api.airvisual.com/v2/countries?key={key}'

resp = requests.get(url)
```

```python
paises = resp.json()['data']
x = False

for i in paises:
    if i == {'country': 'España'}:
        x = True
        
if x == True:
    print('España está en AirVisual')
else:
    print('España NO está en AirVisual')
```

```python
#Muestra las ciudades de Andorra que tienen una estacion y agregalas a un dataframe
```

```python
pais = 'Andorra'
url2 = f'http://api.airvisual.com/v2/states?country={pais}&key={key}'

resp2 = requests.get(url2)
```

```python
todos = resp2.json()['data']
ciudades = []

for i in todos:
    ciudades.append(i['state'])
    
```

```python
df = pd.DataFrame(ciudades, columns=['Ciudad'])
df
```

```python
#Muestra las ciudades del estado de california (en EEUU) que tienen estaciones de AirVisual
```

```python
pais2 = 'USA'
estado = 'California'
url3 = f'http://api.airvisual.com/v2/cities?state={estado}&country={pais2}&key={key}'

resp3 = requests.get(url3)
```
